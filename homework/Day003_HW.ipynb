### 比較一下範例檔案中的「File I/O」與「xmltodict」讀出來的內容有什麼差異


```python
'''
Your Code
'''
# File I/O:
fh = open("./64_72hr_CH.xml", "r", encoding='utf-8')
#print(xml)

xml = fh.read()
fh.close()
print(type(xml))

# 解析檔案內容
import xmltodict
data = dict(xmltodict.parse(xml))
print(type(data))
```

#### [Answer]: File I/O : 字串格式。 xmltodict: 字典格式。
