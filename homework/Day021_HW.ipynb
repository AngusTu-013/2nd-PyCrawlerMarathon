## 作業目標

根據範例：

1. 取出今天所有的新聞
2. 取出現在時間兩小時內的新聞

## 1. 取出今天所有的新聞

```python
'''
Your Code
'''
from selenium import webdriver
from bs4 import BeautifulSoup
import time

from datetime import datetime
dateTimeObj = datetime.now()
#print(dateTimeObj)
#print(dateTimeObj.year, '/', dateTimeObj.month, '/', dateTimeObj.day)
date_now = '{}/0{}/{}'.format(str(dateTimeObj.year), str(dateTimeObj.month),str(dateTimeObj.day))
print(date_now)

browser = webdriver.Chrome(executable_path='chromedriver')
browser.get("https://www.ettoday.net/news/news-list.htm")

# 每隔兩秒鐘自動往下滑
for i in range(15):
    time.sleep(2)
    browser.execute_script("window.scrollTo(0, 10000);")
    
# 取得資料，丟到 BeautifulSoup 解析
html_source = browser.page_source
soup = BeautifulSoup(html_source, "html5lib")

for d in soup.find(class_="part_list_2").find_all('h3'):
    date = d.find(class_="date").text.split(" ")[0]
    #print(date)
    if date != date_now:
        break
    else:
        print(d.find(class_="date").text, d.find_all('a')[-1].text)
```

## 2. 取出現在時間兩小時內的新聞

```python
'''
Your Code
'''
from datetime import datetime
#dateTimeObj = datetime.now()
dateTimeObj = datetime.today()  # Get timezone naive now
#print(dateTimeObj)
t1 = dateTimeObj.timestamp()
print(dateTimeObj, ": ",t1)
#print(dateTimeObj.year, '/', dateTimeObj.month, '/', dateTimeObj.day)
#print(dateTimeObj.hour, ':', dateTimeObj.minute, ':', dateTimeObj.second, '.', dateTimeObj.microsecond)

# Timestamp 轉成 時間
#question_time = '1584496191'
#question_time_date = datetime.fromtimestamp(int(question_time))
#print(question_time_date)
# 時間 轉成 Timestamp:
#t = datetime.datetime(2011, 10, 21, 0, 0)
#time.mktime(t.timetuple())


for d in soup.find(class_="part_list_2").find_all('h3'):
    date = d.find(class_="date").text.split(" ")[0]
    time = d.find(class_="date").text.split(" ")[1]
    news_date = date + ' ' + time + ':00'
    #print(news_date)
    t2 = datetime.strptime(news_date, "%Y/%m/%d %H:%M:%S")#total_seconds()
    #print(t2.timestamp())
    interval_time = (t1 - t2.timestamp())
    #print(interval_time)

    if interval_time > 7200:
        break
    else:
        print(d.find(class_="date").text, d.find_all('a')[-1].text)
```

## 3. 根據範例，取出三天前下午三點到五點的新聞

```python
'''
Your Code
'''
from datetime import datetime
dateTimeObj = datetime.today()  # Get timezone naive now
print(dateTimeObj)
#t1 = dateTimeObj.timestamp()
today_am00 = str(dateTimeObj.year) + '/0' + str(dateTimeObj.month) + '/' + str(dateTimeObj.day) + ' ' + "00:00:00"
print(today_am00)
t1 = datetime.strptime(today_am00, "%Y/%m/%d %H:%M:%S").timestamp()
print(dateTimeObj, ": ",t1)
#print(dateTimeObj.year, '/', dateTimeObj.month, '/', dateTimeObj.day)
#print(dateTimeObj.hour, ':', dateTimeObj.minute, ':', dateTimeObj.second, '.', dateTimeObj.microsecond)

# Timestamp 轉成 時間
#question_time = '1584496191'
#question_time_date = datetime.fromtimestamp(int(question_time))
#print(question_time_date)
# 時間 轉成 Timestamp:
#t = datetime.datetime(2011, 10, 21, 0, 0)
#time.mktime(t.timetuple())


for d in soup.find(class_="part_list_2").find_all('h3'):
    date = d.find(class_="date").text.split(" ")[0]
    time = d.find(class_="date").text.split(" ")[1]
    news_date = date + ' ' + time + ':00'
    #print(date)
    t2 = datetime.strptime(news_date, "%Y/%m/%d %H:%M:%S")#total_seconds()
    #print(t2.timestamp())
    interval_time = (t1 - t2.timestamp())
    print(interval_time)

    #today00:00:00 -> (48hr+ 8hr)(00:00:00 - 17:00:00) ~ (48hr+10hr)
    if (interval_time > ((48+8)*60*60)) and (interval_time <= ((48+10)*60*60)):
        break
    else:
        print(d.find(class_="date").text, d.find_all('a')[-1].text)
```
